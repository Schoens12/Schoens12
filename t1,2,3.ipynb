```python
import random
import numpy as np
from math import sqrt,atan
from statistics import mean, variance
```

```python
def f(z):
    return sqrt(atan(z))
N=100000
l=[]
for i in range(N):
    u=random.random()
    l.append(f(u))
print('La integral estimada es:',mean(l))
print('Con una varianza de:',variance(l))    
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from math import sin,pi
```

```python
def buffon2(n):
    lx = []
    ly = []
    l = 0.6 #longitud de la aguja
    a = 1 #de ancho de líneas paralelas
    k = 0;#Registra el número de veces que la aguja se cruza con la línea paralela
    for i in range(n):
        y = random.uniform(0, a/2)# Obedece la distrinucion uniforme en [0, a/2] para generar n numeros aleatorios
        theta = random.uniform(0, pi)#Obedece la distribución uniforme en [0, pi] para generar n números aleatoriamente
        if y < (l/2)*sin(theta):
            k = k + 1
            lx.append(k)
            ly.append((2*l*n)/(a*k))
    Pi = (2*l*n)/(a*k)
    e = np.absolute((Pi - np.pi)/np.pi)
    print("pi =", Pi)
    print("error =", e)
    
    fig = plt.figure(1, figsize=(8, 8))        
    plt.xlabel("Iteraciones")
    plt.ylabel("Estimación de $\pi$")
    plt.grid()
    plt.plot(lx, ly, color="blue", markersize=1)
    plt.show()
    print("visualizando las ultimas 100 iteraciones:")
    fig = plt.figure(1, figsize=(8, 8))        
    plt.xlabel("Iteraciones")
    plt.ylabel("Estimación de $\pi$")
    plt.grid()
    plt.plot(lx[k-100:k], ly[k-100:k], color="blue", markersize=1)
    plt.show()
```

```python
buffon2(1000000)
```

```python
from random import random, gauss
from scipy import stats
import matplotlib.pyplot as plt
def normunif(n):
    U=[]
    for j in range(n):
        ui=0
        for i in range(12):
            ui += random()
        U.append(ui - 6)
    return U
```

```python
uu=normunif(10000)
print(stats.kstest(uu, 'norm'))
plt.hist(uu,density=1,color='grey',edgecolor='black')
plt.show()
nn=[]
for j in range(10000):
    nn.append(gauss(0,1))
print(stats.kstest(nn, 'norm'))
plt.hist(nn,density=1,color='grey',edgecolor='black')
plt.show()
```
